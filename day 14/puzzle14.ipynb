```python
def rotate_90(matrix):
    transposed = list(zip(*matrix[::-1]))
    rotated_matrix = [list(row) for row in transposed]

    return rotated_matrix
```

```python
def tilt_field(field):
    new_field = copy.deepcopy(field)
    dict = {}
    length = len(lines)

    for ix, el in enumerate(field[0]):
        dict[ix] = []
    
    for line_ix, line in enumerate(field):
        for el_ix, el in enumerate(line):
            #print(el)
            if dict[el_ix] == [] and el == "O":
                continue

            elif el == "#":
                dict[el_ix] = []
            elif el == ".":
                dict[el_ix].append(length - line_ix)
            elif el == "O":
                dict[el_ix].append(length - line_ix)
                new_field[line_ix][el_ix] = "."
                new_field[length-dict[el_ix][0]][el_ix] = "O"
                dict[el_ix] = dict[el_ix][1:]
               
    return new_field
```

```python
def count_load(field):
    total_sum = 0
    length = len(lines)

    
    for line_ix, line in enumerate(field):
        for el_ix, el in enumerate(line):
            if el == "O":
                total_sum += length - line_ix
               
    return total_sum
```

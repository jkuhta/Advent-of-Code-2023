```python
with open('input.txt') as f:
   lines = f.readlines() 
```

```python
plan = []
for line in lines:
    _, _, ins = line.split('\n')[0].split(" ")
    num = int(ins[2:-2], 16)
    dir = int(ins[-2:-1])
    plan.append((dir, num))
    
print(plan)
```

```python
def next_el(x,y, dir):
    if dir == 3:
        return (x-1, y)
    if dir == 1:
        return (x+1, y)
    if dir == 0:
        return (x, y+1)
    if dir == 2:
        return (x, y-1)
    
```

```python
cur = (0,0)
elements = []

for dir, num in plan:
    for step in range(num):
        cur = next_el(cur[0], cur[1], dir)
        elements.append(cur)
```

```python
def count_enclosed_tiles(vertices):
    # SHOELACE FORMULA
    n = len(vertices)

    if vertices[0] != vertices[-1]:
        vertices.append(vertices[0])

    area = 0.0

    for i in range(n):
        area += (vertices[i][0] * vertices[i + 1][1]) - (vertices[i + 1][0] * vertices[i][1])

    area = abs(area) / 2.0
    
    # PICK'S THEOREM
    enclosed_count = abs(n / 2 - 1 - area)
    return int(enclosed_count + n)
    
```

```python
lava = count_enclosed_tiles(elements)

print("-- PART 2 --")
print("CUBIC METERS OF LAVA:", lava)
```
